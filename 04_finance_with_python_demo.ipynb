## Most Simple Financial Market

**_Two dates only (today, tomorrow), two traded assets and TWO future states only with equal probability._**

### Traded Financial Assets

```python
!git clone https://github.com/tpq-classes/finance_with_python.git
import sys
sys.path.append('finance_with_python')
```

```python
import numpy as np
```

```python
# risky stock -- future payoff
S = np.array((20, 5))
```

```python
# risky stock -- current price
S0 = 10
```

```python
# risk-less bond -- future payoff
B = np.array((11, 11))
```

```python
# risk-less bond -- current price
B0 = 10
```

### The Market 

```python
# market price vector
M0 = np.array((S0, B0))
```

```python
# market payoff matrix
M = np.array((S, B)).T
M
```

### Expected Utility Maximization (I)

```python
from scipy.optimize import minimize
```

```python
w = 10  # initial wealth
```

```python
def u(x):
    return np.sqrt(x)
```

```python
p = 0.5
```

```python
P = np.array((p, 1-p))
```

```python
def EU(phi):
    c = np.dot(M, phi)
    return np.dot(u(c), P)
```

```python
phi = np.array((0.25, 0.5))
```

```python
EU(phi)
```

```python
bnds = ((0, np.inf), (0, np.inf))
```

Budget constraint:

$$ M_0 \cdot \phi \leq w $$

```python
cons = {'type': 'eq', 'fun': lambda phi: np.dot(M0, phi) - w}
```

```python
opt = minimize(lambda phi: -EU(phi), (1, 1), bounds=bnds, constraints=cons)
opt
```

```python
-opt['fun']  # maximal expected utility (under budget constraint)
```

```python
phi = opt['x']  # optimal portfolio (under budget constraint)
phi
```

```python
np.dot(M0, phi)
```

### Expected Utility Maximization (II)

```python
M0 = np.array((1, S0, B0))
```

```python
kappa = 0.75
```

```python
def EU(phi):
    c0 = phi[0]
    c = np.dot(M, phi[1:])
    return u(c0) + kappa * np.dot(u(c), P)
```

```python
phi = np.array((3, 0.5, 0.5))
```

```python
phi[1:]
```

```python
EU(phi)
```

```python
bnds = 3 * [(0, np.inf)]
bnds
```

```python
cons = {'type': 'eq', 'fun': lambda phi: np.dot(M0, phi) - w}
```

```python
opt = minimize(lambda phi: -EU(phi), (1, 1, 1), bounds=bnds, constraints=cons)
opt
```

```python
phi = opt['x']
```

```python
np.dot(M0, phi)
```

```python
np.dot(M, phi[1:])
```

## Simple Financial Market

**_Two dates only (today, tomorrow), two traded assets and THREE future states only with equal probability._**

### Traded Financial Assets

```python
S = np.array((20, 10, 5))
B = np.array((11, 11, 11))
S0 = 10
B0 = 10
```

### The Market 

```python
M0 = np.array((S0, B0))
M = np.array((S, B)).T
M
```

### Expected Utility Maximization (I)

```python
w = 15  # initial wealth
```

```python
def u(x):
    return np.sqrt(x)
```

```python
P = np.array((1 / 3, 1 / 3, 1 / 3))
```

```python
def EU(phi):
    c = np.dot(M, phi)
    return np.dot(u(c), P)
```

```python
phi = np.array((0.25, 0.5))  # example portfolio
```

```python
EU(phi)
```

```python
bnds = ((0, np.inf), (0, np.inf))
```

```python
cons = {'type': 'eq', 'fun': lambda phi: np.dot(M0, phi) - w}
```

```python
opt = minimize(lambda phi: -EU(phi), (1, 1), bounds=bnds, constraints=cons)
opt
```

```python
-opt['fun']  # maximal expected utility (under budget constraint)
```

```python
phi = opt['x']  # optimal portfolio (under budget constraint)
phi
```

```python
np.dot(M0, phi)
```

```python
np.dot(M, phi)
```

```python
M0 = np.array((1, S0, B0))
```

```python
def EU(phi):
    c0 = phi[0]
    c = np.dot(M, phi[1:])
    return u(c0) + kappa * np.dot(u(c), P)
```

```python
phi = np.array((3, 0.5, 0.5))
```

```python
phi[1:]
```

```python
EU(phi)
```

```python
bnds = 3 * [(0, np.inf)]
bnds
```

```python
cons = {'type': 'eq', 'fun': lambda phi: np.dot(M0, phi) - w}
```

```python
phi = opt['x']
```
