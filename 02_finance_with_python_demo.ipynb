## Most Simple Financial Market

**_Two dates only (today, tomorrow), two traded assets and TWO future states only with equal probability._**

### Traded Financial Assets

```python
!git clone https://github.com/tpq-classes/finance_with_python.git
import sys
sys.path.append('finance_with_python')
```

```python
import numpy as np
np.set_printoptions(suppress=True)
```

```python
# risky stock -- future payoff
S = np.array((20, 5))
```

```python
# risky stock -- current price
S0 = 10
```

```python
# risk-less bond -- future payoff
B = np.array((11, 11))
```

```python
# risk-less bond -- current price
B0 = 10
```

### The Market 

```python
# market price vector
M0 = np.array((S0, B0))
```

```python
# market payoff matrix
M = np.array((S, B)).T
M
```

### A European Call Option 

```python
# contingent claim -- future payoff
K = 14.5
C = np.maximum(S - K, 0)
C
```

## Expected Returns

```python
def ret(x0, x):
    return (x / x0 - 1).mean()
```

```python
rS = ret(S0, S)
rS
```

```python
rB = ret(B0, B)
rB
```

## Discounting &mdash; Wrong Way

```python
# stock expected return
C.mean() / (1 + rS)
```

```python
# bond expected return
C.mean() / (1 + rB)
```

## Replication &mdash; Linear Algebra

```python
# solving system of linear equations
phi = np.linalg.solve(M, C)
phi
```

```python
np.dot(M, phi)
```

```python
np.dot(M, phi) == C
```

```python
C0 = np.dot(M0, phi)
C0  # arbitrage free value of the option
```

## Discounting &mdash; Correct Way

```python
# expected return of option
rC = ret(C0, C)
rC
```

```python
C.mean() / (1 + rC)
```

## Volatility &mdash; Standard deviation

```python
# volatility ...
def vol(x0, x):
    return (x / x0 - 1).std()
```

```python
# ... of stock
vol(S0, S)
```

```python
# ... of bond
vol(B0, B)
```

```python
# ... of option
vol(C0, C)
```

## Approximation &mdash; OLS Regression

```python
phi = np.linalg.lstsq(M, C, rcond=None)[0]
phi
```

```python
C0 = np.dot(M0, phi)
C0
```

## Neural Nets &mdash; Learning Algorithm

<img src="http://hilpisch.com/images/neural_net.png">

```python
# intial weights
w = np.random.random(2)
w
```

```python
# layer one (output)
l1 = np.dot(M, w)
l1
```

```python
# deltas
d = l1 - C
d
```

```python
# mean squared error
(d ** 2).mean()
```

```python
# learning rate
alpha = 0.01
```

```python
# update values
u = alpha * d
u
```

```python
# updating weights
# w = w - u
w -= u
w
```

```python
# putting together
l1 = np.dot(M, w)
d = l1 - C
(d ** 2).mean()
```

```python
l1
```

```python
# running a loop
alpha = 0.025
w = np.random.random(2)
for i in range(200):
    l1 = np.dot(M, w)
    d = l1 - C
    u = alpha * d
    w -= u
    if i % 10 == 0:
        print((d ** 2).mean())
```

```python
l1
```

```python
d
```

## Martingale Pricing &mdash; Probabilistic Approach

```python
# making the stock price a martingale
def E(P):
    return np.dot(S, P) / (1 + rB)
```

```python
p = 0.5
```

```python
P = np.array((p, 1 - p))
P
```

```python
E(P)
```

```python
E(P) == S0
```

```python
from scipy.optimize import minimize
```

```python
bnds = ((0, 1), (0, 1))
```

```python
cons = {'type': 'eq', 'fun': lambda Q: Q.sum() - 1}
```

```python
def error(Q):
    return (E(Q) - S0) ** 2
```

```python
opt = minimize(error, P, bounds=bnds, constraints=cons)
```

```python
opt
```

```python
Q = opt['x'].round(6)
Q
```

```python
E(Q)
```

```python
E(Q) == S0  # Q makes the stock price process a martingale (!)
```

```python
np.dot(C, Q) / (1 + rB)  # risk-neutral valuation
```
