## Super-Replication

```python
C1 = np.array((1, 0, 0))
C1
```

```python
1 / B[1][0] * B[1] >= C1
```

```python
1 / B[1][0] * B[0]
```

```python
def V(phi, t):
    return phi[0] * B[t] + phi[1] * S[t]
```

```python
phi = np.array((0.04, 0.03))
```

```python
V(phi, 0)
```

```python
V(phi, 1)
```

```python
from scipy.optimize import minimize
```

```python
cons = ({'type': 'ineq', 'fun': lambda phi: V(phi, 1) - C1})
```

```python
res = minimize(lambda phi: V(phi, 0),
               (0.01, 0.01),
               method='SLSQP',
               constraints=cons)
```

```python
res
```

```python
V(res['x'], 0)
```

```python
V(res['x'], 1).round(5)
```

## Approximative Replication

```python
M = np.array((B[1], S[1])).T
```

```python
M
```

```python
reg = np.linalg.lstsq(M, C1, rcond=-1)
```

```python
reg
# (array,
#  array,
#  int,
#  array)
```

```python
V(reg[0], 0)
```

```python
V(reg[0], 1)
```

```python
V(reg[0], 1) - C1
```

```python
np.mean((V(reg[0], 1) - C1) ** 2)
```

## Capital Market Line

```python
mu_S = 7 / 6 - 1
```

```python
mu_S
```

```python
sigma_S = (S[1] / S[0]).std()
```

```python
sigma_S
```

```python
s = np.linspace(-2, 2, 25)
```

```python
b = (1 - s)
```

```python
i = 0.1
```

```python
mu = b * i + s * mu_S
```

```python
sigma = np.abs(s * sigma_S)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(sigma, mu)
plt.xlabel('$\\sigma$')
plt.ylabel('$\\mu$')
```

## Capital Asset Pricing Model

```python
beta = np.linspace(0, 2, 25)
```

```python
mu = i + beta * (mu_S - i)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(beta, mu, label='security market line')
plt.xlabel('$\\beta$')
plt.ylabel('$\\mu$')
plt.ylim(0, 0.25)
plt.plot(1, mu_S, 'ro', label='market portfolio')
plt.legend(loc=0);
```
